```python
# Import libraries necessary for this part of the project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Allows the use of display() for DataFrames
from IPython.display import display 

# Pretty display for notebooks
%matplotlib inline
```

```python
# Load the Santander Customer Satisfacations data from a CSV file

data = pd.read_csv('Data/Santander_Customer_Satisfaction.csv') 
```

```python
# 1 - Split the data columns by into features and target
# 2 - exclude the (ID) column, since that, it will not include to the model

target = data[['TARGET']].copy()

features = data.drop(['ID','TARGET'], axis=1).copy()

#target = target.rename(columns={'':'T'})

features.shape
```

## 1-  Exclude Constant Features

```python
# Remove constant variables that always have one unique value because it's not affect the results

constant_columns = ['ind_var2_0', 'ind_var2', 'ind_var27_0', 'ind_var28_0', 'ind_var28', 'ind_var27',\
                   'ind_var41', 'ind_var46_0', 'ind_var46', 'num_var27_0', 'num_var28_0', 'num_var28',\
                   'num_var27', 'num_var41', 'num_var46_0', 'num_var46', 'saldo_var28', 'saldo_var27',\
                   'saldo_var41', 'saldo_var46', 'imp_amort_var18_hace3', 'imp_amort_var34_hace3',\
                   'imp_reemb_var13_hace3', 'imp_reemb_var33_hace3', 'imp_trasp_var17_out_hace3',\
                   'imp_trasp_var33_out_hace3', 'num_var2_0_ult1', 'num_var2_ult1', 'num_reemb_var13_hace3',\
                   'num_reemb_var33_hace3', 'num_trasp_var17_out_hace3', 'num_trasp_var33_out_hace3',\
                   'saldo_var2_ult1', 'saldo_medio_var13_medio_hace3']

features = features.drop(constant_columns, axis = 1)

features.shape
```

## 2- Correlated Features 

```python
# Remove features that represented by other features (have a high correlation with other features)

# 1- Identify Highly Correlated Features
corr_matrix = features.corr().abs()

# 2- Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# 3- Find index of features that have correlation greater than 0.95 wit other features
corr_features = [column for column in upper.columns if any(upper[column] > 0.95)]

# 4- remove high correlated features
features = features.drop(corr_features, axis = 1)

features.shape
```

## 3- Scaling the Features 

Since that the range of values of features varies widely, I have to standardize it using the logarithm transformation.

```python
# Scale the data by natural logarithm
log_features = np.log(features)
```

## 4-  Feature Transformation

### Implementing the PCA

As that the principal component analysis ( PCA ) is usually used to minimize the dataset dimensions space (number of features), so after applying it to the dataset, it will be found the compound combinations of features in a way that has the best description of customers.

```python
def pca_results(good_data, pca):
    '''
    Create a DataFrame of the PCA results
    Includes dimension feature weights and explained variance
    Visualizes the PCA results
    '''

    # Dimension indexing
    dimensions = dimensions = ['Dimension {}'.format(i) for i in range(1,len(pca.components_)+1)]

    # PCA components
    components = pd.DataFrame(np.round(pca.components_, 4), columns = list(good_data.keys()))
    components.index = dimensions

    # PCA explained variance
    ratios = pca.explained_variance_ratio_.reshape(len(pca.components_), 1)
    variance_ratios = pd.DataFrame(np.round(ratios, 4), columns = ['Explained Variance'])
    variance_ratios.index = dimensions

    # Create a bar plot visualization
    #fig, ax = plt.subplots(figsize = (14,8))

    # Plot the feature weights as a function of the components
    #components.plot(ax = ax, kind = 'bar');
    #ax.set_ylabel("Feature Weights")
    #ax.set_xticklabels(dimensions, rotation=0)


    # Display the explained variance ratios
    #for i, ev in enumerate(pca.explained_variance_ratio_):
        #ax.text(i-0.40, ax.get_ylim()[1] + 0.05, "Explained Variance\n          %.4f"%(ev))

    # Return a concatenated DataFrame
    return pd.concat([variance_ratios, components], axis = 1)
```

```python
# Apply PCA by fitting the good data with the same number of dimensions as features

pca = PCA(n_components=12, random_state=0)
pca.fit(features)

# Generate PCA results plot
#pca_results = vs1.pca_results(features, pca)
pca_results = pca_results(features, pca)
#display(pca_results)
#print(pca_results['Explained Variance'].cumsum())
```

```python
display(pca_results)
```

```python
print(pca_results['Explained Variance'].cumsum())
```

```python
# Transform the data using the PCA fit above

reduced_data = pca.transform(features)

reduced_data.shape
```

```python
# Display sample data after applying PCA transformation in 14th dimensions

cols = ['Dimension 1','Dimension 2','Dimension 3', 'Dimension 4','Dimension 5',\
        'Dimension 6','Dimension 7', 'Dimension 8','Dimension 9', 'Dimension 10',\
        'Dimension 11','Dimension 12']

reduced_features = pd.DataFrame(np.round(reduced_data, 4), columns = cols)

display(reduced_features)
```

```python
processed_data = pd.concat([reduced_features,target], axis=1, \
                           join_axes=[reduced_features.index])


display(processed_data)
```
